```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5, 6, 19, 93, 19, 32, 94, 192]
y = [4, 51, 12, 19, 42, 18, 84, 842]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
def myfunc(x):
    return slope * x + intercept
```

```python
model = list(map(myfunc, x))
```

```python
plt.scatter(x, y)
plt.plot(x, y)

plt.show()
```
